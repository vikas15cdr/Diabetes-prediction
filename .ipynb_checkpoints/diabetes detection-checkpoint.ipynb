```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
df['Glucose']=df['Glucose'].replace(0,df['Glucose'].median())
```

```python
df['BloodPressure']=df['BloodPressure'].replace(0,df['BloodPressure'].median())
```

```python
df['SkinThickness']=df['SkinThickness'].replace(0,df['SkinThickness'].median())
```

```python
df['Insulin']=df['Insulin'].replace(0,df['Insulin'].median())
```

```python
df['Pregnancies']=df['Pregnancies'].replace(0,df['Pregnancies'].median())
```

```python
df['BMI']=df['BMI'].replace(0,df['BMI'].mean().round())
```

```python
x = df.iloc[:,:-1]
y = df['Outcome']
```

```python
x.describe()
```

```python
fig, ax = plt.subplots(figsize = (12, 6))
sns.boxplot(data = x, ax=ax)
plt.savefig('boxPlot.jpg')
```

```python
sns.distplot(x.Pregnancies)
```

```python
cols = ['BloodPressure','BMI','Glucose']
for y in cols:
    ub=x[y].mean() + 3* x[y].std()
    lb=x[y].mean() - 3* x[y].std()
    x[y]=np.where(x[y]>=ub,ub,x[y])
    x[y]=np.where(x[y]<=lb,lb,x[y])
    print(lb,ub)
```

```python
cols = ['Pregnancies', 'SkinThickness', 'Insulin', 'DiabetesPedigreeFunction', 'Age']
for col in cols:
    Q1 = x[col].quantile(0.25)
    Q3 = x[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    x[col]= np.where(x[col] <= lower_bound,lower_bound,x[col])
    x[col]= np.where(x[col] >= upper_bound,upper_bound,x[col])
    print(lower_bound,upper_bound)
```

```python
x.describe()
```

```python
fig, ax = plt.subplots(figsize = (12, 6))
sns.boxplot(data = x, ax=ax)
plt.savefig('boxPlot.jpg')
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled_df = pd.DataFrame(x_scaled)
cols = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']
x_scaled_df.columns = cols
```

```python
fig, ax = plt.subplots(figsize = (12, 6))
sns.boxplot(data = x_scaled_df, ax=ax)
plt.savefig('boxPlot.jpg')
```

```python
import scipy.stats as stat
import pylab
```

```python
def plot_data(df,feature):
    plt.figure(figsize=(12,6))
    plt.subplot(1,2,1)
    df[feature].hist()
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.show()
```

```python
x_scaled_df['Glucose_boxcox'],parameters=stat.boxcox(x_scaled_df['Glucose']+3)
plot_data(x_scaled_df,'Glucose_boxcox')
```

```python
x_scaled_df['BloodPressure_boxcox'],parameters=stat.boxcox(x_scaled_df['BloodPressure']+4)
plot_data(x_scaled_df,'BloodPressure_boxcox')
```

```python
x_scaled_df['SkinThickness_boxcox'],parameters=stat.boxcox(x_scaled_df['SkinThickness']+3)
plot_data(x_scaled_df,'SkinThickness_boxcox')
```

```python
x_scaled_df['BMI_boxcox'],parameters=stat.boxcox(x_scaled_df['BMI']+3)
plot_data(x_scaled_df,'BMI_boxcox')
```

```python

x_scaled_df['DiabetesPedigreeFunction_boxcox'],parameters=stat.boxcox(x_scaled_df['DiabetesPedigreeFunction']+2)
plot_data(x_scaled_df,'DiabetesPedigreeFunction_boxcox')
```

```python

x_scaled_df['Age_boxcox'],parameters=stat.boxcox(x_scaled_df['Age']+2)
plot_data(x_scaled_df,'Age_boxcox')
```

```python
x_transformed = x_scaled_df[['Pregnancies','Insulin','Glucose_boxcox',
       'BloodPressure_boxcox', 'SkinThickness_boxcox', 'BMI_boxcox',
       'DiabetesPedigreeFunction_boxcox', 'Age_boxcox']].copy()
```

```python
y_transformed = pd.DataFrame(df.Outcome)
y_transformed.columns = ['Outcome']
```

```python
y_transformed.head()
```

```python
y_transformed['Outcome'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_transformed, y_transformed, test_size=0.33, random_state=42)

```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
x_trained_resample,y_trained_resample = smote.fit_resample(X_train,y_train)
```

```python
y_trained_resample.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
classification = LogisticRegression()
classification.fit(x_trained_resample,y_trained_resample)
```

```python
y_pred = classification.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)

from sklearn.metrics import classification_report
target_names = ['Non-Diabetic', 'Diabetic']
print(classification_report(y_test, y_pred, target_names=target_names))
```
